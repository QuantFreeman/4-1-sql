# 1.1 Movies Database

Take the movies dataset and turn it into a single `sqlite` database. It should have one table for each csv file in the movies dataset

```python
import sqlite3
import pandas as pd
import glob
```

```python
g = glob.glob('../Data/movie_data/*.csv')
g
```

```python


conn = sqlite3.connect('../Data/movie_data/movies.sqlite')
c = conn.cursor()

for csv in g:
    print(csv)
    df = pd.read_csv(csv)
    n = csv.split('\\')[1]
    n = n.split('.')[0]
    
    df.to_sql(
        name=n,
        con=conn, 
        schema=None, 
        if_exists='replace', 
        index=True 
    )
```

**1.2.1** Use a single query to pull the original title of movies with a budget above $5m

**1.2.2** Use a query to pull the english-language films with the word `war` in their title

**1.2.3** Left join the average ratings from the `ratings` table onto the `movies_metadata` table, so you can have a relation between budget and rating. Hint: use a subquery.

```python
pd.read_sql_query(
"""
SELECT *
FROM movies_metadata
""", 
    con=conn
)
```

```python
pd.read_sql_query(
"""
SELECT original_title
FROM movies_metadata
WHERE budget > 5000000
""", 
    con=conn
)
```

```python
pd.read_sql_query(
"""
SELECT original_title
FROM movies_metadata
WHERE original_title LIKE '% war %'
AND original_language LIKE 'en'
""", 
    con=conn
)
```

```python
#  join ratings + meta
#  have a relation between budget and rating. Hint: use a subquer

pd.read_sql_query(
"""
Select *
FROM movies_metadata as m
LEFT JOIN ratings as r
    ON m.id = r.movieId
""", 
    con=conn
)
```

# 2. Baseball Database

The [Baseball Database](http://www.seanlahman.com/baseball-archive/statistics/) has an sqlite version. Download it for these exercises.

**2.1** Which player has had the most homeruns?

**2.2** Is there a relation between how many homeruns a player has made in a year and his salary that year? Pull both colums together in a single query



```python
"""
From readme:
2.2 batting Table
2.15 salaries table
"""
url = '../Data/baseballdatabank-master/core/lahmansbaseballdb.sqlite'
conn = sqlite3.connect(url)
```

```python
# Which player has had the most homeruns?
pd.read_sql_query(
"""
SELECT playerID as player, HR as homerun
FROM batting
ORDER BY HR desc
""", 
    con=conn
)
```

```python
# Is there a relation between how many homeruns a player has made in a year and his salary that year? 
# Pull both colums together in a single query

df = pd.read_sql_query(
"""
Select b.playerID, s.salary, b.HR
FROM batting as b
LEFT JOIN salaries as s
    ON b.playerID = s.playerID
""", 
    con=conn
)
```
